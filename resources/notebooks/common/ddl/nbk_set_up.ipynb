### Fetch Job Parameter

```python
environment = dbutils.widgets.get("env")
```

### Create Catalog If Not Exists

```python
spark.sql(f"CREATE CATALOG IF NOT EXISTS {environment}_oc_dab")
print(f"Catalog Created: {environment}_oc_dab")
```

### Create Bronze Schema If Not Exists

```python
spark.sql(f"CREATE SCHEMA IF NOT EXISTS {environment}_oc_dab.{environment}_bronze")
print(f"Bronze Schema Created: {environment}_oc_dab.{environment}_bronze")
```

### Create Silver Schema If Not Exists

```python
spark.sql(f"CREATE SCHEMA IF NOT EXISTS {environment}_oc_dab.{environment}_silver")
print(f"Silver Schema Created: {environment}_oc_dab.{environment}_silver")
```

### Create Gold Schema If Not Exists

```python
spark.sql(f"CREATE SCHEMA IF NOT EXISTS {environment}_oc_dab.{environment}_gold")
print(f"Gold Schema Created: {environment}_oc_dab.{environment}_gold")
```

### Create Bronze Volume Root Path

```python
spark.sql(f"""
            CREATE VOLUME IF NOT EXISTS {environment}_oc_dab.{environment}_bronze.volume_bronze
            COMMENT 'This is the Bronze Layer Managed Volume';
          """)
print(f"Bronze Layer Managed Volume Created: /Volumes/{environment}_oc_dab/{environment}_bronze/volume_bronze/")
```
